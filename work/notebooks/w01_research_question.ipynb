## 2. The question: decision, action, cost of a wrong call

### Research question

**Which content pages should editors review first to reduce the risk of losing search visibility?**

### Decision

Determine which pages should be prioritized for content review and refresh.

### Who acts?

Website editors or content managers.

### Action

Editors review the highest-ranked pages and decide whether to refresh, expand, rewrite, merge, or monitor the content.

### Cost of a wrong recommendation

A **false positive** means an editor spends time reviewing a page that did not actually need attention.

A **false negative** means an important declining page is missed, which may lead to continued loss of search traffic and missed business opportunities.

### Why this is not just "train a model"

The goal is not simply to predict which pages are declining. The goal is to improve a business decision by helping editors prioritize their limited time. The model is only a tool for producing a better review queue.
